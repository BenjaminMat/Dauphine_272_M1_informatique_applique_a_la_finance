## Exercise : The Vigenère Cipher

The Vigenère cipher is a substitution encryption system that creates many possible encryptions for the same letter depending on its position in the sentence. This encryption method works using a keyword. Here, we will consider the word **"clef"**. The principle is as follows:

* Transform the keyword into a sequence of integers corresponding to the position of each letter of the word in the alphabet. Here, this gives: *clef* → 3 12 5 6

* To encrypt a sentence, shift: the first letter by 3 letters, the second by 12, the third by 5, the fourth by 6, then repeat: the fifth letter is shifted by 3, the sixth by 12, etc.
