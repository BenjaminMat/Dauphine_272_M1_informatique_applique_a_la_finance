Write a function that takes a string as an argument and returns the number of vowels it contains.


```python
def count_vowels(text: str) -> int:
    vowels = "aeiouy"
    return sum(1 for char in text.lower() if char in vowels)

def count_vowels_(text: str) -> int:
    vowels = "aeiouy"
    count_ = 0
    for char in text.lower():
        if char in vowels: 
            count_ = count_ + 1
    return count_

# tests
print(count_vowels("Bonjour"))      # 3 (o, o, u)
print(count_vowels_("Bonjour"))     # same output than count_vowels but different implementation
print(count_vowels("ChatGPT"))      # 1 (a)
print(count_vowels("PYTHON"))       # 1 (o)
```

#### 2 
Write a function that takes a string as an argument and returns `True` if it contains more vowels than consonants, and `False` otherwise.


```python
def has_more_vowels(text: str) -> bool:
    vowels = "aeiouy"
    text = text.lower()
    nb_vowels = sum(1 for char in text if char in vowels)
    nb_consonants = sum(1 for char in text if char.isalpha() and char not in vowels)
    return nb_vowels > nb_consonants

# test
print(has_more_vowels("Bonjour"))    # False (3 vowels, 4 consonants)
print(has_more_vowels("oui"))        # True  (3 vowels, 0 consonants)
print(has_more_vowels("Python"))     # False (1 vowel, 5 consonants)
```

## Exercise : Match ends

Given a list of strings, return the count of the number of strings where the string length is 2 or more and the first and last chars of the string are the same.
Note: python does not have a ++ operator, but += works.

```python
def match_ends(words_list: list[str]):
    count = 0
    for word in words_list:
        if len(word) >= 2 and word[0] == word[-1]:
            count += 1
    return count

def match_ends_alternative(words_list: list[str]) -> int:
    list_with_matching_criteria = [element for element in words_list if len(element) > 2 and element[0] == element[-1]]
    return len(list_with_matching_criteria)

example_list = ['kayak', 'ok', 'hat', 'equity', 'rotator']
print(match_ends(example_list))
print(match_ends_alternative(example_list))
```

```python
def somme_entier(n: int) -> int:
    if n <= 0:
        return 0
    return n + somme_entier(n - 1)

# test
print(somme_entier(0))   # 0
print(somme_entier(4))   # 10
print(somme_entier(10))  # 55
```

## Exercise : Factorial

Write a function that takes a natural number as an argument and returns its factorial.


```python
def factorielle(N : int) -> int:
    resultat = 1
    for i in range(1, N+1):
        resultat *= i
    return resultat

def factorielle_alternative(N: int) -> int:
    result = 1
    [result*i for i in range(1, N+1)]
    return result
```

Write an equivalent recursive way to do it (the function must call itself).


```python
def factorielle_recursive(N : int) -> int:
    if N == 0:
        return 1
    else:
        return N * factorielle_recursive(N-1)
```

You will use the following matrix to test your function:


```python
M = [[.1, .1], [2, 1], [.1, .5], [3, 1.5]]
```

```python
from math import sqrt

def euclidian_norm(point1 : list, point2 : list =[0,0]) -> float :
    vector = [point1[i] - point2[i] for i in range(len(point1))]
    return sqrt(sum(list(map(lambda x : x**2, vector))))

#map(func, iterable) applies func to each item in iterable, producing an iterator of results.
print(sorted(M, key = euclidian_norm))
```

#### 2
Write a function that takes as input a matrix (such as the one from the previous exercise) and a vector (a list of size 2, in this case) and returns the row of the matrix corresponding to the point closest to the given vector.

```python
def nearest_row(matrix: list[list], vector: list) -> list:
    for row in matrix:
        if len(row) != len(vector):
            raise ValueError("all rows must have the same length as vector")
    return min(matrix, key=lambda row: euclidian_norm(row, vector)) #min returns the original element for which key(element) is minimal

vector = [1,1]
print(nearest_row(M, vector))
```

## Exercise : Magical Square

In mathematics, a magic square of order $n$ is composed of $n^2$ strictly positive integers, written in the form of a square array. These numbers are arranged so that their sums over each row, each column, and each main diagonal are equal.

Write a function that takes as input a square (in the form of a list of lists of the same size, where row $i$ represents the $i^{th}$ row of the square) and returns `True` if it is a magic square, and `False` otherwise.


```python
from typing import List
from typing import List

def is_magic_square(square: List[List[int]]) -> bool:
    n = len(square)
    magic_sum = sum(square[0])

    diag1 = 0
    diag2 = 0
    # Loop over rows and columns
    for i in range(n):
        row_sum = 0
        col_sum = 0
        for j in range(n):
            row_sum += square[i][j]
            col_sum += square[j][i]
        # If the sum of the row or column is different from the magic sum, return False
        if row_sum != magic_sum or col_sum != magic_sum:
            return False
        # Compute diagonals
        diag1 += square[i][i]
        diag2 += square[i][n - i - 1]

    # Check that the diagonals are equal to the magic sum
    if diag1 != magic_sum or diag2 != magic_sum:
        return False
    # If all checks pass, it is a magic square
    return True

# Example usage
magic_square = [
    [8, 1, 6],
    [3, 5, 7],
    [4, 9, 2]
]

non_magic_square = [
    [8, 1, 6],
    [3, 5, 7],
    [4, 9, 3]
]

# Test both squares
print(is_magic_square(magic_square))      # Should return True
print(is_magic_square(non_magic_square))  # Should return False
```

## Exercise : Word frequency analysis

Create a function that takes a text as input and returns a dictionary containing:

* Words as keys
* A tuple as values containing (number of occurrences, list of positions)

Ignore case and punctuation.


Example :

```python
# expected return for => analyze_text("Le chat noir, le chien noir et le chat blanc")
{'Le': (3, [0, 3, 7]),
 'chat': (2, [1, 8]),
 'noir': (2, [2, 5]),
 'chien': (1, [4]),
 'et': (1, [6]),
 'blanc': (1, [9])}
```

```python
def analyze_text(text):
    text = text.lower()
    for p in ',.!?;:':
        text = text.replace(p, '')
    result = {}
    position = 0
    for word in text.split():
        if word not in result:
            result[word] = (1, [position])
        else:
            count, positions = result[word]
            result[word] = (count + 1, positions + [position])
        position += 1
    return result

print(analyze_text("Le chat noir, le chien noir et le chat blanc"))
```

## Exercise : Student grades management

Create a function that manages student grades by subject.
The function takes a list of tuples (student, subject, grade) and returns a dictionary organized by student with the average grade per subject.


Example:

```python
grades = [
    ("Alice", "Physique", 10),
    ("Alice", "Maths", 15),
    ("Bob", "Maths", 12),
    ("Alice", "Maths", 13),
    ("Bob", "Physique", 9),
    ("Alice", "Physique", 16),
    ("Bob", "Physique", 14),
    ("Bob", "Maths", 18)
]

```

```python
def get_student_averages(notes):
    results = {}    
    for student, subject, grade in notes:
        if student not in results:
            results[student] = {}    
        if subject not in results[student]:
            results[student][subject] = []
        results[student][subject].append(grade)
        
    for student in results:
        for subject in results[student]:
            grades = results[student][subject]
            results[student][subject] = sum(grades) / len(grades)
            
    return results

print(get_student_averages(grades))
```
