## Exerise : Vowels

#### 1
Write a function that takes a string as an argument and returns the number of vowels it contains.


```python
def count_vowels(text: str) -> int:

    vowels = "aeiouy"
    return sum(1 for char in text.lower() if char in vowels)

# tests
print(count_vowels("Bonjour"))      # 3 (o, o, u)
print(count_vowels("ChatGPT"))      # 1 (a)
print(count_vowels("PYTHON"))       # 1 (o)
```

#### 2 
Write a function that takes a string as an argument and returns `True` if it contains more vowels than consonants, and `False` otherwise.


```python
def has_more_vowels(text: str) -> bool:

    vowels = "aeiouy"
    text = text.lower()
    nb_vowels = sum(1 for char in text if char in vowels)
    nb_consonants = sum(1 for char in text if char.isalpha() and char not in vowels)
    return nb_vowels > nb_consonants

# test
print(has_more_vowels("Bonjour"))    # False (3 vowels, 4 consonants)
print(has_more_vowels("oui"))        # True  (3 vowels, 0 consonants)
print(has_more_vowels("Python"))     # False (1 vowel, 5 consonants)
```

## Exercise : Match ends

Given a list of strings, return the count of the number of strings where the string length is 2 or more and the first and last chars of the string are the same.
Note: python does not have a ++ operator, but += works.

```python
def match_ends(words_list):
    count = 0
    for word in words_list:
        if len(word) >= 2 and word[0] == word[-1]:
            count += 1
    return count
```

Write a function that takes a natural number as an argument and returns its factorial.


```python
def factorielle(N : int) -> int:
    resultat = 1
    for i in range(1, N+1):
        resultat *= i
    return resultat
```

Write an equivalent recursive way to do it (the function must call itself).


```python
def factorielle_recursive(N : int) -> int:
    if N == 0:
        return 1
    else:
        return N * factorielle_recursive(N-1)
```

## Exercise : Magical Square

In mathematics, a magic square of order $n$ is composed of $n^2$ strictly positive integers, written in the form of a square array. These numbers are arranged so that their sums over each row, each column, and each main diagonal are equal.

Write a function that takes as input a square (in the form of a list of lists of the same size, where row $i$ represents the $i^{th}$ row of the square) and returns `True` if it is a magic square, and `False` otherwise.


```python
from typing import List

def est_carre_magique(carre: List[List[int]]) -> bool:
    n = len(carre)
    
    somme_magique = sum(carre[0])
    
    somme_diag1 = 0
    somme_diag2 = 0
    
    # loop over rows and columns
    for i in range(n):
        somme_ligne = 0
        somme_colonne = 0
        
        for j in range(n):
            somme_ligne += carre[i][j]
            somme_colonne += carre[j][i]
        
        # If the sum of the row or column is different from the magic sum, return False
        if somme_ligne != somme_magique or somme_colonne != somme_magique:
            return False
        
        # Compute diagonals
        somme_diag1 += carre[i][i]
        somme_diag2 += carre[i][n - i - 1]
    
    # Check that the diagonals are equal to the magic sum
    if somme_diag1 != somme_magique or somme_diag2 != somme_magique:
        return False
    
    # If all checks pass, it is a magic square
    return True

# Example usage
carre_magique = [
    [8, 1, 6],
    [3, 5, 7],
    [4, 9, 2]
]

carre_non_magique = [
    [8, 1, 6],
    [3, 5, 7],
    [4, 9, 3]
]

# Test both squares
print(est_carre_magique(carre_magique))      # Should return True
print(est_carre_magique(carre_non_magique))  # Should return False
```

## Exercise : Student grades management

Create a function that manages student grades by subject.
The function takes a list of tuples (student, subject, grade) and returns a dictionary organized by student with the average grade per subject.


```python
def get_student_averages(notes):

    results = {}
    
    for student, subject, grade in notes:
        # If the student does not exist yet, create their dictionary
        if student not in results:
            results[student] = {}
            
        # If the subject does not exist yet for this student, create their list of grades
        if subject not in results[student]:
            results[student][subject] = []
            
        # Add the grade
        results[student][subject].append(grade)
    
    # Compute averages
    for student in results:
        for subject in results[student]:
            grades = results[student][subject]
            results[student][subject] = sum(grades) / len(grades)
            
    return results
```

```python
notes = [
    ("Alice", "Physique", 10),
    ("Alice", "Maths", 15),
    ("Bob", "Maths", 12),
    ("Alice", "Maths", 13),
    ("Bob", "Physique", 9),
    ("Alice", "Physique", 16),
    ("Bob", "Physique", 14),
    ("Bob", "Maths", 18)
]

get_student_averages(notes)
```

### Exercise : Shopping cart


```python
def analyze_purchases(achats):
    # Initialize counters
    article_counts = {}
    total_par_client = {}
    articles_par_client = {}
    
    # Counting and totals
    for client, article, quantite, prix in achats:
        # Count items
        if article not in article_counts:
            article_counts[article] = 0
        article_counts[article] += quantite
        
        # Total per client
        if client not in total_par_client:
            total_par_client[client] = 0
        total_par_client[client] += quantite * prix
        
        # Items per client
        if client not in articles_par_client:
            articles_par_client[client] = set()
        articles_par_client[client].add(article)
    
    # Create suggestions
    suggestions = {}
    all_articles = set(article_counts.keys())
    
    for client in articles_par_client:
        # Suggestions are the articles the client has not purchased
        suggestions[client] = all_articles - articles_par_client[client]
    
    # Sort top-selling items
    top_articles = sorted(article_counts.items(), key=lambda x: x[1], reverse=True)
    
    return {
        "top_articles": top_articles,
        "total_par_client": total_par_client,
        "articles_par_client": articles_par_client,
        "suggestions": suggestions
    }
```

```python
achats = [
    ("Client1", "Pommes", 3, 1.5),  # (customers, items, quantity, unit_price)
    ("Client1", "Poires", 2, 2.0),
    ("Client2", "Pommes", 2, 1.5),
    ("Client2", "Bananes", 4, 1.0),
    ("Client3", "Poires", 1, 2.0),
    ("Client3", "Bananes", 6, 1.0)
]
analyze_purchases(achats)
```

## Exerise : Returns analysis

```python
import statistics
from typing import List, Tuple, Dict

def analyser_rendements(donnees_actions: List[Tuple[str, List[float]]]) -> Tuple[str, List[str], float]:
    """
    Takes a list of tuples (stock_name, [monthly_returns]) and returns:
    1. The stock with the highest average return
    2. A list of stocks whose returns have always been positive
    3. The median return of all stocks combined
    """
    rendements_moyens = []
    actions_positives = []
    tous_rendements = []

    for nom, rendements in donnees_actions:
        rendement_moyen = sum(rendements) / len(rendements)
        rendements_moyens.append((nom, rendement_moyen))
        
        if all(r > 0 for r in rendements):
            actions_positives.append(nom)
        
        tous_rendements.extend(rendements)

    action_meilleur_rendement = max(rendements_moyens, key=lambda x: x[1])[0]
    rendement_median = statistics.median(tous_rendements)

    return action_meilleur_rendement, actions_positives, rendement_median

# Example data
donnees = [
    ("AAPL", [2.5, -1.3, 3.2, 1.1, -0.7, 2.3]),
    ("GOOGL", [1.8, 2.1, -0.9, 3.3, 1.7, 0.6]),
    ("MSFT", [1.2, 1.9, 2.3, 1.1, 1.6, 1.4]),
    ("AMZN", [3.1, -2.2, 1.7, 2.4, -1.1, 3.6])
]

# Tests
resultat = analyser_rendements(donnees)
print("Results of the return analysis:")
print(f"Stock with the highest average return: {resultat[0]}")
print(f"Stocks with always positive returns: {resultat[1]}")
print(f"Overall median return: {resultat[2]:.2f}")
```

## Exercise : Option pricing with Black–Scholes and computation of the Greeks


```python
import math
from scipy.stats import norm
from typing import List, Tuple, Dict

def black_scholes_merton(S: float, K: float, T: float, r: float, sigma: float, option_type: str = 'call') -> Dict[str, float]:
    """
    Computes the price of an option and its Greeks according to the Black–Scholes–Merton model.
    
    :param S: Current price of the underlying asset
    :param K: Strike price of the option
    :param T: Time to expiration (in years)
    :param r: Risk-free interest rate (annual)
    :param sigma: Volatility of the underlying asset
    :param option_type: Option type ('call' or 'put')
    :return: A dictionary containing the option price and its Greeks
    """
    d1 = (math.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * math.sqrt(T))
    d2 = d1 - sigma * math.sqrt(T)

    if option_type == 'call':
        price = S * norm.cdf(d1) - K * math.exp(-r * T) * norm.cdf(d2)
        delta = norm.cdf(d1)
    else:  # put
        price = K * math.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
        delta = norm.cdf(d1) - 1

    gamma = norm.pdf(d1) / (S * sigma * math.sqrt(T))
    vega = S * norm.pdf(d1) * math.sqrt(T)
    theta = -S * norm.pdf(d1) * sigma / (2 * math.sqrt(T)) - r * K * math.exp(-r * T) * norm.cdf(d2)
    rho = K * T * math.exp(-r * T) * norm.cdf(d2)

    if option_type == 'put':
        theta += r * K * math.exp(-r * T)
        rho = -K * T * math.exp(-r * T) * norm.cdf(-d2)

    return {
        'price': price,
        'delta': delta,
        'gamma': gamma,
        'vega': vega,
        'theta': theta,
        'rho': rho
    }

# Example usage for Exercise 2
S = 100   # Current asset price
K = 100   # Strike price
T = 1     # Time to expiration (1 year)
r = 0.05  # Risk-free rate (5%)
sigma = 0.2  # Volatility (20%)

call_results = black_scholes_merton(S, K, T, r, sigma, 'call')
put_results = black_scholes_merton(S, K, T, r, sigma, 'put')

print("\nResults for the Call Option:")
for key, value in call_results.items():
    print(f"{key.capitalize()}: {value:.4f}")

print("\nResults for the Put Option:")
for key, value in put_results.items():
    print(f"{key.capitalize()}: {value:.4f}")
```

## Exercise : Simplified Monte Carlo Simulation

Basic calculations to give intuition on the asymmetry of downside returns:

200 days at **+1%** and 200 days at **-1%** on a portfolio with an initial value of **100** lead to a total return of **-2%**.


```python
100*0.99**200*1.01**200
```

Expected returns using exponential returns for a distribution centered at **0%** with **20%** annualized volatility:


```python
math.exp(0.2**2/2)
```

```python
import random
import math
from typing import List, Tuple, Dict, Literal
import matplotlib.pyplot as plt

def simuler_prix_action(prix_initial: float, volatilite: float, jours: int,
                        simulations: int, simul_type : Literal["continuous", "discrete"]) -> List[float]:
    """
    Performs a Monte Carlo simulation of a stock price.
    
    :param prix_initial: Initial stock price
    :param volatilite: Annual volatility of the stock
    :param jours: Number of days to simulate
    :param simulations: Number of simulations to run
    :param simul_type: Simulation type ("continuous" or "discrete")
    :return: List of final prices for each simulation
    """
    resultats = []
    volatilite_quotidienne = volatilite / math.sqrt(252)  # 252 trading days per year
    
    for _ in range(simulations):
        prix = prix_initial
        for _ in range(jours):
            rendement = random.gauss(0, volatilite_quotidienne)
#             print(rendement)
            if simul_type == "discrete":
                prix *= (1 + rendement)
            elif simul_type == "continuous":
                prix *= math.exp(rendement)
            
        resultats.append(prix)
    
    return resultats

# Example usage
prix_initial = 100
volatilite = 0.2
jours = 252  # One trading year
simulations = 1000

resultats = simuler_prix_action(prix_initial, volatilite, jours, simulations, "continuous")
print(f"Average price after {jours} days: {sum(resultats) / len(resultats)}")
print(f"Minimum price: {min(resultats)}")
print(f"Maximum price: {max(resultats)}")
plt.hist(resultats, bins=40)
plt.show()
```

With the continuous method, the average price is close to 102, which is completely logical because we are working here with cumulative returns of the form

$$
R_{\text{annual}} = \prod_{t=1}^{252} (1 + r_t) - 1
$$

which we approximate using the Taylor expansion of the exponential function of $x$ around 0, giving the following approximation:

$$
R_{\text{annual}} \approx \exp\left(\sum_{t=1}^{252} r_t\right) - 1.
$$

Since returns are Gaussian in this example, we know the form of the expectation of the exponential of a random variable
$X \sim \mathcal{N}(\mu, \sigma^2)$:

$$
\mathbb{E}[\exp(X)] = \exp\left(\mu + \frac{\sigma^2}{2}\right).
$$

In our case $\mu = 0$ and $\sigma^2 = 0.2^2$, which gives an expected annual return of:

$$
\exp\!\left(\tfrac{0.2^2}{2}\right) - 1 = 2.02\%.
$$

When using standard geometric returns, the average return indeed tends toward 0. The use of exponential returns introduces a positive bias due to volatility:

$$
\mathbb{E}[\exp(x)] > \exp(\mathbb{E}[x])
$$

(Jensen’s inequality). The higher the frequency of returns, the closer the continuous form is to reality.


## Exercise : The Vigenère Cipher

The Vigenère cipher is a substitution encryption system that creates many possible encryptions for the same letter depending on its position in the sentence. This encryption method works using a keyword. Here, we will consider the word **"clef"**. The principle is as follows:

* Transform the keyword into a sequence of integers corresponding to the position of each letter of the word in the alphabet. Here, this gives: *clef* → 3 12 5 6

* To encrypt a sentence, shift: the first letter by 3 letters, the second by 12, the third by 5, the fourth by 6, then repeat: the fifth letter is shifted by 3, the sixth by 12, etc.

**Hint:** use Python’s `ord()` function, and remember to handle case sensitivity and non-alphanumeric characters.


#### 1 
Write a function that takes a sentence as an argument and returns its encrypted version according to the Vigenère cipher with the keyword **"clef"**.

```python
ord('C') - ord('A')
```

```python
def vigenere_encrypt_clef(text: str) -> str:
    """
    Encrypt "text" using Vigenère cipher with fixed key "clef"
    - Preserves case
    - Non-letters are left unchanged
    """
    # key is 'clef', hence shifts are [3, 12, 5, 6] with a=1,...,z=26
    shifts = [3, 12, 5, 6]
    m = len(shifts)
    out_chars = []
    j = 0  # index over letters only

    for ch in text:
        if ch.isalpha():
            k = shifts[j % m]
            if ch.isupper():
                base = ord('A')
                out_chars.append(chr(base + ((ord(ch) - base + k) % 26)))
            else:
                base = ord('a')
                out_chars.append(chr(base + ((ord(ch) - base + k) % 26)))
            j += 1
        else:
            out_chars.append(ch)
    return "".join(out_chars)

# Quick checks
print(vigenere_encrypt_clef("Bonjour a tous"))
print(vigenere_encrypt_clef("Hello Everyone"))
```

Write a function that takes a sentence encrypted with the Vigenère cipher using the keyword **"clef"** and returns its decrypted version.


```python
def vigenere_decrypt_clef(ciphertext: str) -> str:
    """
    Decrypt `ciphertext` encrypted with Vigenère cipher using fixed key 'clef'.
    - Preserves case.
    - Non-letters are left unchanged.
    """
    shifts = [3, 12, 5, 6]
    m = len(shifts)
    out_chars = []
    j = 0  # index over letters only

    for ch in ciphertext:
        if ch.isalpha():
            k = shifts[j % m]
            if ch.isupper():
                base = ord('A')
                out_chars.append(chr(base + ((ord(ch) - base - k) % 26)))
            else:
                base = ord('a')
                out_chars.append(chr(base + ((ord(ch) - base - k) % 26)))
            j += 1
        else:
            out_chars.append(ch)
    return "".join(out_chars)

# Quick checks (round trip)
plain = "Bonjour a tous"
enc = vigenere_encrypt_clef(plain)
dec = vigenere_decrypt_clef(enc)
print(enc)
print(dec)
```
